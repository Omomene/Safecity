```python
# 3 Helper: Load Excel safely
def load_excel_safe(file_path: Path):
    """
    Load an Excel file (.xls or .xlsx) and return the first non-empty sheet as a DataFrame.
    """
    ext = file_path.suffix.lower()
    
    if ext == ".xlsx":
        engine = "openpyxl"
    elif ext == ".xls":
        engine = "xlrd"
    else:
        raise ValueError(f"Unsupported Excel format: {file_path.name}")
    
    xls = pd.ExcelFile(file_path, engine=engine)
    print(f"Sheets in {file_path.name}: {xls.sheet_names}")
    
    # Find first non-empty sheet
    for sheet in xls.sheet_names:
        df = pd.read_excel(file_path, sheet_name=sheet, engine=engine)
        if not df.empty:
            print(f"Using sheet: {sheet}")
            return df
    raise ValueError(f"No non-empty sheets found in {file_path.name}")

```

```python
# 4 Load Data
crimes = load_excel_safe(CRIME_RAW_FILE)
population = pd.read_excel(POPULATION_RAW_FILE, sheet_name="2023", engine="xlrd")


# Quick check
print("Crimes columns:", crimes.columns.tolist())
print("Population columns:", population.columns.tolist())
display(crimes.head())
display(population.head())

```

```python
def load_population_safe(file_path):
    xls = pd.ExcelFile(file_path, engine="xlrd")
    for sheet in xls.sheet_names:
        df = pd.read_excel(file_path, sheet_name=sheet, engine="xlrd")
        # Keep only numeric columns
        numeric_cols = df.select_dtypes(include="number")
        if not numeric_cols.empty:
            print(f"Using sheet: {sheet}")
            return df
    raise ValueError("No numeric population data found")
```

```python
# 5 Cleaning Functions
def clean_crime_data(df: pd.DataFrame) -> pd.DataFrame:
    """Clean crime data: standardize columns, drop duplicates, fill missing values."""
    df = df.copy()
    df.columns = [c.lower().strip() for c in df.columns]
    df = df.drop_duplicates()
    df = df.fillna(0)
    return df

def clean_population_data(df: pd.DataFrame) -> pd.DataFrame:
    """Clean population data: standardize columns and drop duplicates."""
    df = df.copy()
    df.columns = [c.lower().strip() for c in df.columns]
    df = df.drop_duplicates()
    return df

# 6 Apply Cleaning
crimes_clean = clean_crime_data(crimes)
population_clean = clean_population_data(population)

# Check cleaned data
display(crimes_clean.head())
display(population_clean.head())

```

```python
# 7 Optional: AI Report
def generate_ai_report(df: pd.DataFrame, description="crime") -> str:
    """
    Generate a short AI report (in French) based on the descriptive stats of the DataFrame.
    """
    summary = df.describe().to_string()
    
    response = completion(
        model="openrouter/meta-llama/llama-3.1-8b-instruct",
        messages=[
            {"role": "system", "content": "You are a data analyst specialized in public safety statistics."},
            {"role": "user", "content": f"Voici les statistiques rsumes des donnes de {description} :\n\n{summary}\n\nRdige un rapport analytique court (en franais) expliquant les principales tendances et anomalies potentielles."}
        ]
    )
    
    return response.choices[0].message.content

# 8 Generate Report (Optional)
# ai_report = generate_ai_report(crimes_clean)
# report_path = REPORTS_DIR / "crime_analysis.txt"
# report_path.write_text(ai_report, encoding="utf-8")
# print(f" AI report generated: {report_path}")
```
